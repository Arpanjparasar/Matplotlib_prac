```python
#question no 1 #Using the file Assessment_data.csv, Showing the marks distribution of User35 in the four subjects. 
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Software Engineering', 'Web Tech', 'Java', 'Algorithm'
sizes = [50.8, 26, 33.29, 55.67]
explode = (0, 0, 0, 0.1)  # only "explode" the 4th slice (i.e Algorithm)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

# Inference for question 1

1. user35 have scored 20.1% in Software Enginerring
2. user35 have scored 30.6% in Web Tech
3. user35 have scored 15.7% in Java
4. user35 have scored 33.6% in Algorithm

```python
# question 4 # Using Survey_data.csv file, count number of times different Life Goals are stated.
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize']=(10,6)
goals=['Manager','Developer','Database Designer','Software Architect']
numbers=[83,293,42,70]
ypos = np.arange(len(goals))

plt.yticks(ypos,goals)
plt.xlabel("Number of Students")
plt.ylabel("GOALS")
plt.title("Goals of Students")
plt.barh(ypos,numbers,align='center', color='blue')
plt.show()
```

# Inference for question 4

1. From the above bar chart its is seen that 83 students wants to be Manager
2. From the above bar chart its is seen that 293 students wants to be Developer which is the highest among the given data
3. From the above bar chart its is seen that 42 students wants to be Database Designer whic is lowest amon the given data
4. From the above bar chart its is seen that 70 students wants to be Software Architect

```python
#question no 3 Using PopulationData.csv Comparing  the yearly population growth rate of India and China.
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
plt.rcParams['figure.figsize']=(10,6)
X=[1.69,1.89,2.06,2.15,2.34,2.32,2.33,2.17,2.00,1.86,1.67,1.47,1.24,1.15,1.13,1.11]
Y=[1.96,1.49,1.90,2.68,2.23,1.54,1.50,1.83,1.13,0.69,.059,0.57,0.54,0.46,0.43,0.39]
x_labels=[1995,1960,1965,1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2016,2017,2018]
plt.xticks(np.arange(len(x_labels)),x_labels,rotation=90)
plt.plot(X,color='Blue')
plt.plot(Y,color='Red')
plt.title("Yearly population growth of India and China")
plt.xlabel("Year")
plt.ylabel("Population")
blue_patch=mpatches.Patch(color='Blue',label='India')
red_patch=mpatches.Patch(color='Red',label='China')
plt.legend(handles=[blue_patch,red_patch])
plt.grid(True)
plt.show()
```
